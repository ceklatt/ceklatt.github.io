```python
# convert M/F to binary (1 for male, 0 for female)
df['Sex'] = np.where(df['Sex'] == 'M', 1, df['Sex'])
df['Sex'] = np.where(df['Sex'] == 'F', 0, df['Sex'])
```

```python
# convert Fatal Y/N to binary (1 for fatal, 0 for non-fatal)
df['Fatal'] = np.where(df['Fatal'] == 'Y', 1, df['Fatal'])
df['Fatal'] = np.where(df['Fatal'] == 'N', 0, df['Fatal'])
```

```python
df
```

```python
x = df[['Sex','Age','Fatal']]
y = df['Type']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state = 5)
knnclassifier = KNeighborsClassifier(n_neighbors = 5)
knnclassifier.fit(x_train,y_train)
y_pred = knnclassifier.predict(x_test)
metrics.accuracy_score(y_test, y_pred)
```

```python
# accuracy is 87.7%
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state = 3)
knnclassifier = KNeighborsClassifier(n_neighbors = 5)
knnclassifier.fit(x_train,y_train)
y_pred = knnclassifier.predict(x_test)
metrics.accuracy_score(y_test, y_pred)
```

```python
# accuracy is 85.9%
```

```python
# KNN
neighbors = np.arange(1, 9)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

for i, k in enumerate(neighbors):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
      
    # Compute traning and test data accuracy
    train_accuracy[i] = knn.score(x_train, y_train)
    test_accuracy[i] = knn.score(x_test, y_test)
```

```python
plt.plot(neighbors, test_accuracy, label = 'Testing dataset Accuracy')
plt.plot(neighbors, train_accuracy, label = 'Training dataset Accuracy')
  
plt.legend()
plt.xlabel('n_neighbors')
plt.ylabel('Accuracy')
plt.show()
```

```python
from scipy.stats import chi2_contingency 
import numpy as np
chisqt = pd.crosstab(df.Sex, df.Fatal, margins=True)
value = np.array([chisqt.iloc[0][0:5].values,
                  chisqt.iloc[1][0:5].values])
print(chi2_contingency(value)[0:3])
```

```python
# 2 degrees of freedom
# p-value is 0.755
# statistical value is 0.56
# p-value is greater than 0.05, so we accept null hypthesis that Sex and fatality are independent of each other
```

```python
chisqt = pd.crosstab(df.Sex, df.Type, margins=True)
value = np.array([chisqt.iloc[0][0:5].values,
                  chisqt.iloc[1][0:5].values])
print(chi2_contingency(value)[0:3])
```
