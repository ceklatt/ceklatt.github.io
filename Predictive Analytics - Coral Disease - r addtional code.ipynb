```r
library(ggplot2)
library(tidyverse)
```

```r
df <- read.csv('Coral_Data.csv')
```

```r
df
```

```r
df1 <- df %>% select(GENUS, DZCODE)
```

```r
df1
```

```r
ggplot(data = df1, aes(x=GENUS,y=DZCODE)) + 
  geom_boxplot()+
  scale_color_brewer(palette="Dark2") + 
  geom_jitter(shape=16, position=position_jitter(0.2))+
  labs(title = 'Coral Disease',
       y='Disease',x='Genus')
```
